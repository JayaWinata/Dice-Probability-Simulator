```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_number = int(input("Number(s) of dice rolled / Experiment: "))
one = two = three = four = five = six = 0
```

```python
for i in range(exp_number):
    random_num = random.randint(1,6)
    if random_num == 1:
        one += 1
    elif random_num == 2:
        two += 1
    elif random_num == 3:
        three += 1
    elif random_num == 4:
        four += 1
    elif random_num == 5:
        five += 1
    else:
        six += 1
```

```python
x = [1,2,3,4,5,6]
event_num = [one,two,three,four,five,six]
y = [one/exp_number,two/exp_number,three/exp_number,four/exp_number,five/exp_number,six/exp_number]
prob = 1/6
info = 'Theoritical Probability: ' + str(prob) + ' (' + (f'{prob*100:.0f}') + ' %)'
```

```python
fig, (plot1,plot2) = plt.subplots(1,2,figsize=(10,4))

plot1.pie(event_num,labels=x,autopct='%.0f%%')
plot2.bar(x,y)
plt.xlabel('Dice Number')
plt.ylabel('Probability')
fig.text(0.05,0.05,info)
fig.show(False)
```

```python
dice_num = [1,2,3,4,5,6]
cumulative_prob = np.cumsum(y).tolist()
for i, v in enumerate(cumulative_prob):
    plt.text(dice_num[i], v, f"{v:.2f}", ha='center', va='bottom')
plt.xlabel('Dice Number')
plt.ylabel('Cumulative Probability')
plt.yticks(np.arange(0,1.05,0.05))
plt.bar(dice_num,cumulative_prob)
plt.show()
```
